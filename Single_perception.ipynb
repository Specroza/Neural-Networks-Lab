```python
# Dataset: [study_hours, attendance_percentage]
X = [
    [2, 50],
    [3, 60],
    [4, 65],
    [5, 70],
    [6, 80],
    [1, 40],
    [7, 90],
    [4, 60],
    [5, 55],
    [6, 65]
]

# PASS (1) if study_hours >= 4 AND attendance >= 65 else FAIL (0)
y_true = [0, 0, 1, 1, 1, 0, 1, 0, 0, 1]
```

```python
w1 = 0.1
w2 = 0.1
b = -5
learning_rate = 0.01
```

```python
def step_function(weighted_sum):
    if weighted_sum >= 1:
        return 1
    else:
        return 0
```

```python
epochs = 30

for epoch in range(epochs):
    error_count = 0

    for i in range(len(X)):
        study = X[i][0]
        attendance = X[i][1] / 100   # normalization

        weighted_sum = (w1 * study) + (w2 * attendance) + b
        prediction = step_function(weighted_sum)

        error = y_true[i] - prediction

        if error != 0:
            error_count += 1
            w1 += learning_rate * error * study
            w2 += learning_rate * error * attendance
            b += learning_rate * error

    print(f"Epoch {epoch+1}, Errors: {error_count}")

    if error_count == 0:
        break
```

```python
correct_predictions = 0

for i in range(len(X)):
    study = X[i][0]
    attendance = X[i][1] / 100   # same normalization

    weighted_sum = (w1 * study) + (w2 * attendance) + b
    prediction = step_function(weighted_sum)

    if prediction == y_true[i]:
        correct_predictions += 1

accuracy = (correct_predictions / len(X)) * 100

print("Accuracy of the model:", accuracy, "%")
```

```python
study = float(input("Enter study hours per day: "))
attendance = float(input("Enter attendance percentage: "))

attendance = attendance / 100

weighted_sum = (w1 * study) + (w2 * attendance) + b
result = step_function(weighted_sum)

if result == 1:
    print("PREDICTION: The student is likely to PASS")
else:
    print("PREDICTION: The student is likely to FAIL")
```

Short Report:
Single Layer Perceptron
1. Dataset Creation
The dataset was created manually using two input features: study hours per day and attendance percentage. A total of ten data points were used. The labels were assigned based on a predefined rule where a student is classified as Pass if the study hours are greater than or equal to four and the attendance percentage is greater than or equal to sixty-five. Otherwise, the student is classified as Fail. This rule was chosen to realistically represent student performance.

2. Learning Rate Selection
The learning rate was chosen as 0.01. This value allows the perceptron to update its weights gradually and prevents instability during training. A small learning rate ensures smooth convergence of the model across multiple training epochs.

3. Model Learning Verification
The learning of the model was verified by training the perceptron for several epochs and observing the reduction in errors. As the number of epochs increased, the number of misclassifications decreased, indicating that the model was learning effectively. After training, the accuracy of the model was calculated by comparing the predicted outputs with the actual labels. The final accuracy achieved was approximately 80 percent, which confirms successful learning.
